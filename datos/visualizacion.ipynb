```python
# Import the required library
from geopy.geocoders import Nominatim

# Initialize Nominatim API
geolocator = Nominatim(user_agent="MyApp")

location = geolocator.geocode("Mogilev")

print("The latitude of the location is: ", location.latitude)
print("The longitude of the location is: ", location.longitude)
```

```python
# Lazo para obtener coordenadas - Grupo 1

destinos = []
for index, row in df_ranking_destinos_1.iterrows():
    location = geolocator.geocode(row['destino'])
    print(index, row['destino'], row['cantidad'], location.latitude, location.longitude)

    destino_fila = {'destino':row['destino'],'cantidad':row['cantidad'], 'latitud':location.latitude, 'longitud':location.longitude}
    destinos.append(destino_fila)
```

```python
# Lazo para obtener coordenadas - Grupo 2

destinos_2 = []
for index, row in df_ranking_destinos_2.iterrows():
    location = geolocator.geocode(row['destino'])
    print(index, row['destino'], row['cantidad'], location.latitude, location.longitude)

    destino_fila = {'destino':row['destino'],'cantidad':row['cantidad'], 'latitud':location.latitude, 'longitud':location.longitude}
    destinos_2.append(destino_fila)
```
